```python
# Create a dictionary to store the sub-column information for each main column
sub_column_info = {}

# Iterate through the columns
for column in df.columns:
    # Check if the column is a tuple (MultiIndex)
    if isinstance(column, tuple):
        # The first part of the tuple is the main column name
        main_column = column[0]
        # The second part is the sub-column name
        sub_column = column[1]
    else:
        # If it's not a tuple, consider it as the main column name
        main_column = column
        sub_column = None

    # If the main column is not in the sub_column_info dictionary, initialize it
    if main_column not in sub_column_info:
        sub_column_info[main_column] = {"Sub-Columns": []}

    # Append the sub-column name to the list (None if it's not a MultiIndex)
    if sub_column is not None:
        sub_column_info[main_column]["Sub-Columns"].append(sub_column)

# Print the main column, sub-columns, and sub-column count
for main_column, info in sub_column_info.items():
    sub_columns = info["Sub-Columns"]
    sub_column_count = len(sub_columns)
    print(f"Main Column: {main_column}")
    print(f"Sub-Columns: {', '.join(sub_columns)}")
    print(f"Sub-Column Count: {sub_column_count}")
    print("\n")
```

## **Data standardization**

```python
# Iterate through columns to show attributes
for column in df.columns:
    print(f"Column: {column}")

    # Get the subcolumn data within the column
    subcolumn_data = df[column]
    unique_values = subcolumn_data.unique()
    unique_count = subcolumn_data.nunique()
    data_type = subcolumn_data.dtypes

    print(f"  Unique Values: {unique_values}")
    print(f"  Number of Unique Values: {unique_count}")
    print(f"  Data Type: {data_type}\n")
```

I noticed that some columns needs to be standardize :

1. **"le travail à la maison"** has 4 unique values : 'oui', 'Oui', 'non' and 'Non'. We should edit our code in order to have 'oui' and 'non'.

2. **"interne/externe"** has 5 unique values : 'interne', 'Interne', 'externe', 'Externe' and 'Extrene'. We should have 'interne' and 'externe'.

3. **"aide aux devoirs"** has 3 unique values : 'oui', 'non' and 'Non'. We should edit our code in order to have 'interne' and 'externe'.

4. **"Public / Privé"** has 2 unique values : 'Public' and 'public'. We should have only 'public'.

5. **"motivation"** has 3 unique values : 'oui', 'Oui' and 'Non'.  We should edit our code in order to have 'oui' and 'non'.


List of columns and subcolumns to standardize :


```python
columns_to_standardize = [
    ('Informations personnelles', 'le travail à la maison'),
    ('Informations personnelles', 'interne / externe'),
    ('Information familiale', 'aide aux devoirs'),
    ("Informations sur l'école et le lieu de résidence", 'Public / Privé'),
    ('Informations psychologiques', 'motivation'),
]
```

Function to standardize text values to lowercase

```python
def standardize_to_lowercase(column):
    df[column] = df[column].str.lower()
```

Apply standardization to the list of columns

```python
for column in columns_to_standardize:
    standardize_to_lowercase(column)
```

New let's check if the problem is solved or not :

```python
# List of columns and subcolumns to check after standardization
columns_to_check = [
    ('Informations personnelles', 'le travail à la maison'),
    ('Informations personnelles', 'interne / externe'),
    ('Information familiale', 'aide aux devoirs'),
    ("Informations sur l'école et le lieu de résidence", 'Public / Privé'),
    ('Informations psychologiques', 'motivation'),
]

# Check and display unique values in the specified columns
for column in columns_to_check:
    unique_values = df[column].unique()
    print(f"Unique values in '{column}': {unique_values}")
```

As we can see, the subcolumn 'interne / externe' has 4 unique values we should edit it :

```python
# Function to standardize text values in the 'interne / externe' subcolumn
def standardize_interne_externe(main_column, sub_column):
    mapping = {
        'interne': 'interne',
        'externe': 'externe',
        'interne ': 'interne',
        'extrene': 'externe',
    }

    df[(main_column, sub_column)] = df[(main_column, sub_column)].map(mapping)

# Apply standardization to the 'interne / externe' subcolumn
standardize_interne_externe('Informations personnelles', 'interne / externe')

# Check and display unique values in the 'interne / externe' subcolumn
unique_values = df[('Informations personnelles', 'interne / externe')].unique()
print(f"Unique values in 'interne / externe': {unique_values}")
```

Also the column **"Age"** contains borth dates not ages so let's change this by converting the column into datetime format and substract it from  the current date :

```python
df[("Informations personnelles", "Age")] = pd.to_datetime(df[("Informations personnelles", "Age")], format="%d/%m/%Y")

# The current date
current_date = datetime.now()

# Calculate the age by subtracting the birthdate from the current date
df[("Informations personnelles", "Age")] = (current_date - df[("Informations personnelles", "Age")]).dt.days // 365
```
